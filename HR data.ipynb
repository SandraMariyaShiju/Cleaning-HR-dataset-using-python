```python
import pandas as pd
df = pd.read_csv('HR Data.csv')
df
```

```python
#no of instances
print("There are",len(df.axes[0]),"instances.")
#no:of variables
print("There are",len(df.axes[1]),"variables.")
#variable names
print('The columns of the dataset are \n', df.columns)
print('The information about the dataset is \n', df.info())
#describing the dataset
print('The description about the dataset is \n',df.describe())
```

```python
# 1. Remove unnecessary columns
columns_to_remove = ['EmployeeCount', 'Over18', 'StandardHours']
df.drop(columns=columns_to_remove, inplace=True)
```

```python
# 2. Rename the columns
new_column_names = {
    'BusinessTravel': 'business_travel',
    'DailyRate': 'daily_rate',
    'DistanceFromHome': 'distance_from_home',
    'EducationField': 'education_field',
    'EmployeeNumber': 'employee_number',
    'EnvironmentSatisfaction': 'environment_satisfaction',
    'HourlyRate': 'hourly_rate',
    'JobInvolvement': 'job_involvement',
    'JobLevel': 'job_level',
    'JobRole': 'job_role',
    'JobSatisfaction': 'job_satisfaction',
    'MaritalStatus': 'marital_status',
    'MonthlyIncome': 'monthly_income',
    'MonthlyRate': 'monthly_rate',
    'NumCompaniesWorked': 'num_companies_worked',
    'PercentSalaryHike': 'percent_salary_hike',
    'PerformanceRating': 'performance_rating',
    'RelationshipSatisfaction': 'relationship_satisfaction',
    'StockOptionLevel': 'stock_option_level',
    'TotalWorkingYears': 'total_working_years',
    'TrainingTimesLastYear': 'training_times_last_year',
    'WorkLifeBalance': 'work_life_balance',
    'YearsAtCompany': 'years_at_company',
    'YearsInCurrentRole': 'years_in_current_role',
    'YearsSinceLastPromotion': 'years_since_last_promotion',
    'YearsWithCurrManager': 'years_with_current_manager'
}
df.rename(columns=new_column_names, inplace=True)
```

```python
#checking for null values
df.isnull().any()
```

```python
# 3. Eliminate redundant entries (duplicates)
df.drop_duplicates(inplace=True)
```

```python
# 4. Sanitize specific columns 
df['Department'] = df['Department'].str.strip()
df['education_field'] = df['education_field'].str.strip()
df['job_role'] = df['job_role'].str.strip()
df['marital_status'] = df['marital_status'].str.strip()
```

```python
# 5. Eliminate NaN values
df.dropna(inplace=True)
```

```python
# Display the cleaned DataFrame
print("\nCleaned DataFrame:")
print(df.head())
```
